# Read original files

```python
# nextclade 
dec_clade = pd.read_csv('/mnt/NAS4/home/jiwon/GISAID_SNU/data_20230918/latest_deceased_nextclade.tsv',delimiter='\t')
hos_clade = pd.read_csv('/mnt/NAS4/home/jiwon/GISAID_SNU/data_20230918/latest_hospitalized_nextclade.tsv',delimiter='\t')
```

```python
# patients info
dec_meta = pd.read_csv('/mnt/NAS4/home/jiwon/GISAID_SNU/data_20230918/latest_deceased_low.metadata.tsv',delimiter='\t')
hos_meta1 = pd.read_csv('/mnt/NAS4/home/jiwon/GISAID_SNU/data_20230918/latest_hospitalized_low_p1.metadata.tsv',delimiter='\t')
hos_meta2 = pd.read_csv('/mnt/NAS4/home/jiwon/GISAID_SNU/data_20230918/latest_hospitalized_low_p2.metadata.tsv',delimiter='\t')
hos_meta3 = pd.read_csv('/mnt/NAS4/home/jiwon/GISAID_SNU/data_20230918/latest_hospitalized_low_p3.metadata.tsv',delimiter='\t')
```

```python
# Concatenate the DataFrames vertically
hos_meta = pd.concat([hos_meta1, hos_meta2, hos_meta3], axis=0)
# Reset the index
hos_meta = hos_meta.reset_index(drop=True)
print(hos_meta.shape)

hos_meta_unique = hos_meta.drop_duplicates()
print(hos_meta_unique.shape)  # check for duplicated data set
```

# merge clade data with patient meta data

```python
merged_dec = pd.merge(dec_meta, dec_clade, how='left', left_on='strain', right_on='seqName')
print(dec_meta.shape, dec_clade.shape, merged_dec.shape)
```

```python
merged_hos = pd.merge(hos_meta, hos_clade, how='left', left_on='strain', right_on='seqName')
print(hos_meta.shape, hos_clade.shape, merged_hos.shape)
```
